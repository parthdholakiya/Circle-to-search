```python
import torch
from torchvision import transforms
from PIL import Image
from googlesearch import search

# Load pre-trained YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5x')

# Function to detect objects in an image using YOLOv5
def detect_objects(image_path):
    # Load image
    img = Image.open(image_path)

    # Perform object detection
    results = model(img)

    # Get detected object names
    object_names = results.names

    # Get bounding boxes and class labels
    boxes = results.xyxy[0].cpu().numpy()
    class_labels = results.xyxy[0][:, -1].cpu().numpy().astype(int)

    # Filter out duplicate class labels
    unique_class_labels = list(set(class_labels))

    # Get object names
    detected_objects = [object_names[label] for label in unique_class_labels]

    return detected_objects

# Function to perform Google search based on detected objects
def google_search(detected_objects):
    search_results = []
    for obj in detected_objects:
        query = obj  # Use detected object as search query
        # Perform Google search
        search_results.extend(search(query, num=5, stop=5, pause=2))

    return search_results

# Example usage
def Circle_to_Search(image_path):
    # Perform object detection
    detected_objects = detect_objects(image_path)

    # Perform Google search based on detected objects
    search_results = google_search(detected_objects)

    # Print search results
    for result in search_results:
        print(result)
```
